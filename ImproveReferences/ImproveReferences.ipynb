```python
with open('bibliography.bib', 'rb') as file:
    library = bibtexparser.load(file)
```

```python
unrecognized_journals = []
for entry in library.entries:
    ID = entry['ID']
    authors = entry['author']
    
    if re.search(r'[A.-Z.]\.[A-Z]', authors):
        print(f'{ID:25}: {"Author":10} : {authors}')
    
    if 'url' in entry and 'doi' in entry:
        del entry['url']
    
    for key in ('number', 'file', 'group', 'month'):
        if key in entry:
            del entry[key]
    
    if not 'doi' in entry:
        print(f'{ID:25}: {"No DOI":10}')
    
    if not 'journal' in entry:
        if entry['ENTRYTYPE'] == 'article':
            print(f'{ID:25}: Missing Journal')
    else:
        journal = entry['journal']
        
        if journal in journals_long:
            entry['journal'] = journals_long[journal]
        elif journal not in journals:
            unrecognized_journals.append(journal)
    
print(f'Unrecognized journals    : {list(set(unrecognized_journals))}')
```

```python
with open('ModifiedReferences.bib', 'w+', encoding='utf8') as file:
    file.write(bibtexparser.dumps(library))
```
